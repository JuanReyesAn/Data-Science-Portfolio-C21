### Basic Queries


#### SELECT (with * and with column names)


```python
query = '''
  SELECT
    name
  FROM
    sqlite_master
  WHERE
    type='table' AND
    name not like "sqlite_%"
;
'''
print(query)
```

```python
# Connect to a sqlite database and load a result set into a data frame
db_con = db.connect("chinook.db")
tables = pd.read_sql_query( query , db_con)

tables

# these are all our tables found within this data frame
```

#### WHERE


```python
# we will chose one of the data tables from our data frame/ lets chose ' tracks'
query = '''
  SELECT *
  FROM employees
  LIMIT 10
'''

employees = pd.read_sql_query(query, db_con)
employees.info()
```

```python
query = '''
  SELECT *
  FROM employees

'''

employees = pd.read_sql_query(query, db_con)
employees
```

```python
#used where to select Lethbridge that in city thats FROM employees that and we SELECT* to see th whole columns of this particular Data Set
query = '''
  SELECT *
  FROM employees
  WHERE city ==  'Lethbridge'
'''

where_e = pd.read_sql_query(query, db_con)
where_e
```

#### AND


```python
# SELECT* full row FROM employees where city is equal to lethbridge an f first name is robert I am basily filtering out things to get what I want
query = ( '''
    SELECT*
    FROM employees
    WHERE city == 'Lethbridge' AND FirstName == 'Robert'
    '''
)
and_e = pd.read_sql_query(query, db_con)
and_e
```

#### OR

```python
# first set the query, assign it to what you would like for it to do
query = ( '''
    SELECT*
    FROM employees
    WHERE city == 'Lethbridge' OR PostalCode == 'T5K 2N1'

    '''
)

or_e = pd.read_sql_query(query, db_con)
or_e
```
