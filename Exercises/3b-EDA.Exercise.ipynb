## Step 6: Univariate plots


Generate at least 3 univariate plots.


```python
eda.hist()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=eda, x='weight', color='lightgreen')
plt.title('Box Plot of Vehicle Weight')
plt.xlabel('Weight')
plt.show()
```

```python
plt.figure(figsize=(10, 2))
sns.countplot(data=eda, x='acceleration', palette='pastel')
plt.title('acceleration of car')
plt.xlabel('acceleration')
plt.ylabel('mph')
plt.show()
```

## Step 7: Bivariate plot (1/2)


Generate a correlation plot.


```python
plt.figure(figsize=(10, 8))

corr_matrix = eda.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title('Correlation Heatmap of Automobile Features', fontsize=16)
plt.show()
```

## Step 8: Bivariate plot (2/2)


Generate a pairplot.


```python
sns.pairplot(eda, hue='origin', palette='viridis')
plt.show()
```
